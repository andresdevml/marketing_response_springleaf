# **Importamos funciones de trabajo**

```python
import pandas as pd

import numpy as np

import joblib

import xgboost as xgb

from sklearn.metrics import roc_auc_score


# conectamos al drive

from google.colab import drive
```

# **Conectamos al drive**

```python

# conectamos al drive

drive.mount('/content/drive')


dir_data_kaggle='./drive/MyDrive/springleaf/data'
```

# **Construimos y entrenamos el modelo**

```python
 # definimos los parametros optimizados

params=joblib.load(dir_data_kaggle+'/xgb_params.sav')

    # preparamos la data

dtrain=xgb.DMatrix(

    data=np.load(dir_data_kaggle+'/df_X_train_prod.npy')  ,

    label=np.load(dir_data_kaggle+'/df_y_train_prod.npy')


                                )

dtest=xgb.DMatrix(

    data=np.load(dir_data_kaggle+'/df_X_test_prod.npy'),

    label=np.load(dir_data_kaggle+'/df_y_test_prod.npy')

                                              )

xgb_clf = xgb.train(
        params=params,
        dtrain=dtrain,
        num_boost_round=5000
                        )
```

# **Evaluamos al modelo**

```python

y_pred=xgb_clf.predict(dtest)

metric=roc_auc_score(np.load(dir_data_kaggle+'/df_y_test_prod.npy'),y_pred)

print(metric) # se logro wey :D
```

# **Nos llevamos los objetos de interes**

```python
# nos llevamos el modelo con joblib

joblib.dump(xgb_clf,dir_data_kaggle+'/xgb_clf.sav')
```

```python
# nos llevamos el modelo del modo recomendado

xgb_clf.save_model(dir_data_kaggle+'/xgb_clf.json')
```
