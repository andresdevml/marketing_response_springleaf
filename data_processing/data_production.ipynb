#**Importamos la data**

```python

# Dfinimos el directorio

dir_data='./drive/MyDrive/springleaf/data'

# obtenemos los indices para dividir la data

mask_length=pd.read_csv(
                filepath_or_buffer=dir_data+'/train.csv'
                        ).shape[0]

mask_index=np.random.rand(mask_length)

train_index=mask_index<=0.9

test_index=mask_index >0.9

```

# **Preparamos el preprocesado**

```python


# importamos la lista de funciones de preprocesado

pre_proc_list=joblib.load(dir_data+'/pre_pipeline.sav')

scaler=joblib.load(dir_data+'/scaler.sav')

# trabajamos sobre la data de entrenamiento

df_data_cat=pd.read_csv(
                filepath_or_buffer=dir_data+'/train.csv').drop(
                                                ['ID', 'target'],
                                                    axis=1).fillna(value=-1).select_dtypes(include=['object'])

# transoformamos dado el encoder


for index, transf in enumerate(pre_proc_list):

  df_data_cat[df_data_cat.columns[index]]=transf.transform(
                     list(df_data_cat[df_data_cat.columns[index]].values))



# llamamos la data numerica

df_data_num=(pd.read_csv(
                filepath_or_buffer=dir_data+'/train.csv')).drop(
                                      ['ID', 'target'],axis=1).fillna(value=-1).select_dtypes(include=['int','float'])

```

```python

# nos llevamos la data

np.save(dir_data+'/df_X_train_prod.npy',scaler.transform( pd.concat(
        [df_data_num[train_index],df_data_cat[train_index]],
        axis=1,
        ignore_index=True)))



np.save(dir_data+'/df_X_test_prod.npy',scaler.transform( pd.concat(
        [df_data_num[test_index],df_data_cat[test_index]],
        axis=1,
        ignore_index=True)))

# nos llevamos los features

# limpiamos la ram

del df_data_cat, df_data_num

# para los targets

np.save(dir_data+'/df_y_train_prod.npy',pd.read_csv(filepath_or_buffer=dir_data+'/train.csv')['target'][train_index])

np.save(dir_data+'/df_y_test_prod.npy',pd.read_csv(filepath_or_buffer=dir_data+'/train.csv')['target'][test_index])


```
